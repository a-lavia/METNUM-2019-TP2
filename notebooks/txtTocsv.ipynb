```python
f = open("outs//out_500.txt","r")
a1 = f.readlines()
f.close()

f = open("outs//out_1000.txt","r")
a2 = f.readlines()
f.close()

f = open("outs//out_1500.txt","r")
a3 = f.readlines()
f.close()

content = a1 + a2 + a3

content
```

```python
vals = []
for c in content:
    aux = c.split('_')
    aux2 = aux[5].split(':')
    aux3 = aux2[0].split('.')
    vals.append((int(aux[2]), int(aux[3]), int(aux[4]), int(aux3[0]), float(aux2[2])))

vals
```

```python
import csv

with open("sinPCA.csv","w") as out:
    csv_out = csv.writer(out)
    csv_out.writerow(["max_features","k","train","test","accuracy"])
    for row in vals:
        csv_out.writerow(row)
```
